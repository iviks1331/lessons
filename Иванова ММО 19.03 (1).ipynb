```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#данные
np.random.seed(42)
X = np.arange(1, 101)
Y = 5 * X + 7 + np.random.uniform(-20, 20, size=100)
```

```python
#потери
def losses(w, b, X, Y):
    return np.mean((w * X + b - Y) ** 2) / 2
```

```python
#градиенты
def compute_gradients(w, b, X, Y):
    n = len(X)
    dw = -np.sum(X * (Y - (w * X + b))) / n
    db = -np.sum(Y - (w * X + b)) / n
    return dw, db
```

```python
#Визуализация сгенерированных данных
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Данные')
plt.title('Генерированные данные')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#градиентный спуск
def gradient_descent(X, Y, lr=0.0001, tol=0.001, max_iters=100000):
    w, b = 0, 0
    for i in range(max_iters):
        dw, db = compute_gradients(w, b, X, Y)
        w -= lr * dw
        b -= lr * db
        if np.linalg.norm([dw, db]) < tol:
            break
    return w, b
```

```python
w_opt, b_opt = gradient_descent(X, Y)
print(f"Оптимальные параметры: w = {w_opt:.5f}, b = {b_opt:.5f}")
```

```python
#градиентный спуск с линией регресии визуализация
plt.scatter(X, Y, label="Данные")
plt.plot(X, w_opt * X + b_opt, color='red', label="Линия регрессии")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
